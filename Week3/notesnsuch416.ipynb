```python
#create a very simple plot

t = arange(0.0, 1.0+0.01, 0.01)
c = cos(4*pi*t)
s = sin(4*pi*t)
```

```python
pi
```

```python
plot(t,c, color="blue", linewidth=5, linestyle="-" ,label="$cos(4(\pi)*t)$")
plot(t,s, color="red",linewidth= 5, linestyle=":", label="$sin(4(\pi)t)$")

xlabel("time(s)")
ylabel("voltage(mV)")
# add title for convenience
title("A simple plot", fontsize=20)
# you can do fontsize for the labels as well
# add gridlines (you can change boldness, density, etc...)
grid(True)
legend(loc="best")
savefig("simple_plot.pdf")
```
