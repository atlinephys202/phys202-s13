```python
x = arange(0,pi+0.001,0.1)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.
plot(x,f(x),lw=3,label="True distribution")
plot(x,F(x),lw=3,label="Cumulative function")
xlim(0,3.5)
ylim(0,1.1)
legend(loc=0)
show()
```

```python
N = 10000
z = random.random_sample(N)
Finv = lambda z: arccos(cos(0)-2*z)
xsamples = Finv(z)
hist(xsamples,50,normed=True,label="Sampled distribution",color="red")
plot(x,f(x),lw=3,label="True distribution",color="blue")
legend(loc=0)
xlim(0,pi)
ylim(0,0.7)
show()
```

```python
x = arange(0,5+0.01,0.01)
f = lambda x: exp(-x)
F = lambda x: -exp(-x)+1
plot(x,f(x),lw=3,label="function")
plot(x,F(x),lw=3,label="CDF")
xlim(0,5.1)
ylim(0,1.1)
legend(loc="center right")
show()
```

```python
N = 10000
z = random.random_sample(N)
Finv = lambda z: -log(1-z)
xsamples = Finv(z)
hist(xsamples,50,normed=True,color="red",label="Sampled distribution")
plot(x,f(x),lw=3,color="blue",label="True distribution")
xlim(0,5)
ylim(0,1)
legend(loc="upper right")
show()
```
