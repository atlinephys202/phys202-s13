```python
x = arange(0,pi+0.001,0.1)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.
plot(x,f(x),lw=3,label="True distribution")
plot(x,F(x),lw=3,label="Cumulative function")
xlim(0,3.5)
ylim(0,1.1)
legend(loc=0)
show()
```

```python
N = 10000
z = random.random_sample(N)
Finv = lambda z: arccos(cos(0)-2*z)
xsamples = Finv(z)
hist(xsamples,50,normed=True,label="Sampled distribution",color="red")
plot(x,f(x),lw=3,label="True distribution",color="blue")
legend(loc=0)
xlim(0,pi)
ylim(0,0.7)
show()
```

```python
def distribute1D(x,prob,N):
    """takes any numerical distribution prob, on the interval defined by the array x, 
    and returns an array of N sampled values that are statistically the same as the input data"""
    normed = cumsum(prob)/cumsum(prob)[len(x)-1]
    z = random.random_sample(N)
    p = searchsorted(normed,z)
    xsamples = x[p]
    return xsamples
```
