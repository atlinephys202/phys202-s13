```python
f = lambda x: x**(1/2.) + sin(x)

x = arange(0.,15.,0.001)

plot(x,f(x),lw=5,color="green")
ylabel('$x^{1/2}+\sin(x)$',fontsize=20)
xlabel("$x$",fontsize=20)
xlim(0.,15.)
x_start = 1.7
x_end = 12.4
y_start = 0.
mask = x[logical_and((x>x_start),(x<x_end))]
y_end = max(f(mask))*1.05
plot([x_start,x_start],[y_start,y_end],color="blue",linestyle="--")
plot([x_end,x_end],[y_start,y_end],color="blue",linestyle="--")
plot([x_start,x_end],[y_end,y_end],color="blue",linestyle="--")

N=1000000

x_samples = (x_end-x_start)*random_sample(N) + x_start
y_samples = y_end*random_sample(N)

newmask = (y_samples < f(x_samples))

plot(x_samples[1:200],y_samples[1:200],"k.")

plot(x_samples[newmask[:200]],y_samples[newmask[:200]],"r.")
show()
```

```python
Integral = (x_end-x_start)*(y_end)*sum(newmask)/newmask.size

print Integral
#(x_end-x_start)*(y_end) is the area of the square
#sum(newmask)/newmask.size   ....... sum(newmask) gives how many dots are underneath the function
#                            ....... newmask.size gives N
#                            ....... dividing gives you the ratio

Ibyhand = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Ibyhand(x_end)-Ibyhand(x_start)

error = (Iexact-Integral)/Iexact*100
print error

```

```python
blah = (x_end-x_start)/N
SMIntegral = sum(f(x_samples))*blah

plot(x,f(x),color="green",linewidth=5)
xlim(0.,15.)
xlabel("(x)",fontsize=20)
ylabel("$x^{1/2}+\sin(x)$",fontsize=20)
plot([x_start,x_start],[y_start,y_end],color="blue",linestyle="--",linewidth=5)
plot([x_end,x_end],[y_start,y_end],color="blue",linestyle="--",linewidth=5)
plot([x_start,x_end],[y_end,y_end],color="blue",linestyle="--",linewidth=5)
lineY=SMIntegral/(x_end-x_start)
plot([x_start,x_end],[lineY,lineY],color="red",linewidth=5)
show()
```
