to check is they're anagrams:
take the input as a string,
break it down into individual letters in list,
rearrange list in alphabetical order,
if lists are the same, return True

```python
def is_anagram(s1,s2):
    ls1 = list(s1)
    ls2 = list(s2)
    if sort(ls1) == sort(ls2):
        print True
    else:
        print False
```
