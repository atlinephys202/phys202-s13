```python
N=1
from random import gauss
Energies = []
while len(Energies) < N:
    Energy = gauss(4000000000,1500000000)
    Energies.append(Energy)
    
print Energies
```

```python
#find velocity (relativistically) from total randomly distributed energy

rest_energy = 105.6583715*(10**6)
rest_energy
```

```python
for i in range(len(Energies)):
    Energies[i] -= rest_energy
    kinetic = Energies
print kinetic
print (str(kinetic))
```

```python
c = 3*(10**8)

v = lambda kin: sqrt(1-((kin**2)/((rest_energy)**2))) * c
print v(kinetic)

###figure out how to make list numbers into acctual usable numbers
```
