Gaussian Distribution
-
$f(x) = \cfrac{1}{\sigma\sqrt{2\pi}}$ $e^{\cfrac{{(x-\mu)}^2}{2\sigma^2}}$

```python
#find velocity (relativistically) from total randomly distributed energy

rest_energy = 105.6583715*(10**6)
#print rest_energy
```

```python
M = len(hit)
from random import gauss
Energies = []
while len(Energies) < M:
    Energy = gauss(4000000000,1500000000)#     gives a total energy to a particle using gaussian distribution
    if Energies>rest_energy:
        Energies.append(Energy)#                   with mean = 4 GeV and sigma = 1.5 GeV
```

```python
for i in range(len(Energies)):
    Energies[i-1] -= rest_energy #subtracts rest energy from total energy to get kinetic energy
    kinetic = Energies
akinetic = array(kinetic) # just an array of Energies
#print akinetic
```

```python
c = 3*(10**8)

v = lambda kin: sqrt(1-((rest_energy**2)/((kin)**2))) * c
vels = v(akinetic)
#print vels
```

```python
# distance travelled will be from int pt to detectors.
for j in hit:
    xdist = int_ptx - xData[j]
    ydist = int_pty - yData[j]
    zdist = 15000 - zData2[j]
    dist = sqrt((xdist**2)+(ydist**2)+(zdist**2))
    #print dist
travel_time = dist / vels
travel_time.sort()
t0 = travel_time[0]
travel_time = dist/vels
#print travel_time
```

```python
delt = random.random(M)
delta = delt*0.05*abs(travel_time - t0)
relative_arrival_times = t0 + delta
#print relative_arrival_times
```
