```python
for i in range(len(Energies)):
    Energies[i-1] -= rest_energy #subtracts rest energy from total energy to get kinetic energy
    kinetic = Energies
akinetic = array(kinetic) # just an array of Energies
#print akinetic
```

Multilateration
=

$0 = xA_m + yB_m + zC_m + D_m$

$A_m = \cfrac{2x_m}{\upsilon\tau_m} - \cfrac{2x_1}{\upsilon\tau_1}$

$B_m = \cfrac{2y_m}{\upsilon\tau_m} - \cfrac{2y_1}{\upsilon\tau_1}$

$C_m = \cfrac{2z_m}{\upsilon\tau_m} - \cfrac{2z_1}{\upsilon\tau_1}$

$D_m = \upsilon\tau_m - \upsilon\tau_1 - \cfrac{x^2_m + y^2_m + z^2_m}{\upsilon\tau_m} + \cfrac{x^2_1 + y^2_1 + z^2_1}{\upsilon\tau_1}$

$A_m(x) + B_m(y) + C_m(z) = -D_m$
